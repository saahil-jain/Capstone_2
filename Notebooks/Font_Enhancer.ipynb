```
% cd /content/Font_Style_Transfer/Code/Generated
% cd Locomo

from google.colab.patches import cv2_imshow
from PIL import Image, ImageFilter
from imutils import contours
from skimage import measure
import numpy as np
import cv2

image_names = os.listdir()
image_names.sort()
generated_images = []
for image_name in image_names:
  if "_O" in image_name:
    generated_images.append(image_name)
generated_images

for generated_image in generated_images:
  image = cv2.imread(generated_image)
  # cv2_imshow(image)
  # image = cv2.addWeighted(image, 2, np.zeros(image.shape, image.dtype), 0, -100)
  # cv2_imshow(image)
  image = cv2.cvtColor(image ,cv2.COLOR_RGB2GRAY)
  _,image = cv2.threshold(image,200,255,cv2.THRESH_BINARY)
  cv2_imshow(image)
  cv2.imwrite(generated_image, image)
```
