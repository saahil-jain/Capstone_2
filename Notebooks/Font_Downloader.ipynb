```
from PIL import Image
import os.path, sys

def crop(path,items):
    for item in items:
        if item == ".DS_Store":
          continue
        fullpath = os.path.join(path,item)  
        print(fullpath)
        if os.path.isfile(fullpath):
            im = Image.open(fullpath)
            width, height = im.size

            min_height = -1
            max_height = height+1
            min_width = -1
            max_width = width+1

            data = im.getdata()

            for i in range(height):
              for j in range(width):
                index = (i*width) + j
                if data[index] != (0,0,0,0):
                  min_height = i
              if min_height!=-1:
                break

            for i in range(height-1,-1,-1):
              for j in range(width):
                index = (i*width) + j
                if data[index] != (0,0,0,0):
                  max_height = i
              if max_height!=height+1:
                break


            for i in range(width):
              for j in range(height):
                index = (j*width) + i
                if data[index] != (0,0,0,0):
                  min_width = i
              if min_width!=-1:
                break

            for i in range(width-1,-1,-1):
              for j in range(height):
                index = (j*width) + i
                if data[index] != (0,0,0,0):
                  max_width = i
              if max_width!=width+1:
                break

            min_height -= 5
            max_height += 5
            min_width -= 5
            max_width += 5


            crop_height = max_height-min_height
            crop_width = max_width-min_width
            center_height = (max_height+min_height)//2
            center_width = (max_width+min_width)//2
            span = (crop_width+1)//2
            if crop_height>crop_width:
              span= (crop_height+1)//2

            min_height = max(center_height-span,0)
            max_height = min(center_height+span, height)
            min_width = max(center_width-span,0)
            max_width = min(center_width+span,width)

            imCrop = im.crop((min_width, min_height, max_width, max_height))
            imCrop.save(fullpath, "PNG", quality=100)


path = "/content"
dirs = ["Quantum"]
for dir in dirs:
  print(dir)
  if dir == ".DS_Store":
    continue
  folders_path = "/content/" + dir
  print(folders_path)
  files = os.listdir(folders_path)
  print(files)
  crop(folders_path,files)
```
